```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```
df=pd.read_csv('/content/matches.csv')
df.head()
```

```
s = pd.value_counts(df['player_of_match'])
s1 = pd.Series({'nunique': len(s), 'unique values': s.index.tolist()})
s.append(s1)
```

```
s = pd.value_counts(df['winner'])
s1 = pd.Series({'nunique': len(s), 'unique values': s.index.tolist()})
s.append(s1)
```

```
s = pd.value_counts(df['season'])
s1 = pd.Series({'nunique': len(s), 'unique values': s.index.tolist()})
s.append(s1)
```

```
print("No. of Matches Played")
df['season'].value_counts()
```

```
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), annot=True)
```

```
sns.countplot(x='season', data=df)
```

```
sns.countplot(x='winner', data=df)
plt.title('Match Winning Teams')
plt.xticks(rotation=90)
```

```
plt.subplots(figsize=(20,10))
sns.countplot(x ='venue',data=df,order=df['venue'].value_counts().index)
plt.title('Indian Premier League Venue',fontsize=35)
plt.xlabel('Stadium',fontsize=15)
plt.ylabel('Matches',fontsize=15)
plt.xticks(rotation=-90,fontsize=15)
```

```
plt.figure(figsize=(16,9))

sns.barplot(x= df.player_of_match.value_counts()[:10].index,y= df.player_of_match.value_counts()[:10])

plt.title('Top player of the match Winners',fontsize=30)
plt.xlabel('Player',fontsize=20)
plt.ylabel('Match',fontsize=20)
```

```
plt.subplots(figsize=(20,10))

sns.countplot('winner', data = df)

plt.title('Champion Teams',fontsize=35)
plt.xlabel('Teams',fontsize=15)
plt.ylabel('Season won',fontsize=15)
plt.xticks(rotation=-90,fontsize=15)
plt.show()
```

```
plt.rcParams['figure.figsize']=70,54
sns.countplot(x=df['venue'], hue='winner', data=df, linewidth=7, palette = "Set2")
plt.ylabel('Number of matches')
plt.xticks(rotation=90)
plt.xlabel('Venue')
plt.title('Venue V/s matches won',size=50, fontweight="bold")
```

```
plt.rcParams['figure.figsize']=70,54
sns.countplot(x=df['season'], hue='winner', data=df, linewidth=7, palette = "Set2")
plt.ylabel('Number of matches')
plt.xticks(rotation=90)
plt.xlabel('year')
plt.title('Season V/s matches won',size=50, fontweight="bold")
```

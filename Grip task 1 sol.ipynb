```
# Importing all libraries required in this notebook
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
```

```
url = "http://bit.ly/w-data"
dataset = pd.read_csv(url)
print("Data imported successfully")

dataset.head(10)
```

```
# Plotting the distribution of scores
dataset.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

```
X=dataset.iloc[:, :1].values
Y=dataset.iloc[:, 1].values
```

```
from sklearn.linear_model import LinearRegression as lg
reg=lg()
reg.fit(X,Y)

Y_pred=reg.predict(X)

plt.scatter(X,Y,color='red')
plt.plot(X,reg.predict(X),color='blue')
plt.title("Study hours V/s percentage (using linear regression)")
plt.xlabel("Study hours")
plt.ylabel("percentage")
plt.show()

print("if a student studies for 9.25 hrs/day his predicted score is",round(reg.predict([[9.5]])[0],2))
```

```
from sklearn import metrics  
print('Mean Absolute Error(for linear regression):', metrics.mean_absolute_error(Y, Y_pred)) 
```

```
from sklearn.ensemble import RandomForestRegressor as rf
lin_reg = rf(n_estimators=200 , random_state=0)
lin_reg.fit(X, Y)
y_pred=lin_reg.predict(X)

plt.scatter(X, Y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title("Study hours V/s percentage (using random forest)")
plt.xlabel("Study hours")
plt.ylabel("percentage")
plt.show()

X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, Y, color = 'red')
plt.plot(X_grid, lin_reg.predict(X_grid), color = 'blue')
plt.title("Study hours V/s percentage (using random forest)")
plt.xlabel("Study hours")
plt.ylabel("percentage")
plt.show()

print("if a student studies for 9.25 hrs/day his predicted score is",round(lin_reg.predict([[9.5]])[0],2))
```

```
from sklearn import metrics  
print('Mean Absolute Error(for random forest):', metrics.mean_absolute_error(Y, y_pred)) 
```
